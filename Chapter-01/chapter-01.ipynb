## 1. 从 RNN 到 Transformer 的演进

### 类比：版本升级

想象"语言理解"技术的版本迭代历程：

![RNN到Transformer演进](images/rnn_to_transformer.png)

*图1：从 RNN 到 Transformer 的演进历程*

| 版本 | 技术 | 年份 | 关键特征 |
|------|------|------|----------|
| v1.0 | RNN | 2014 | 串行处理，速度慢 |
| v2.0 | LSTM / GRU | 2015 | 记忆力更好，但仍是串行 |
| v3.0 | Attention | 2017 | 并行处理！ |
| v4.0 | Transformer | 2017 | 所有现代大模型的基础 |

### v1.0 — RNN：逐字处理

RNN（Recurrent Neural Network）是最早用来处理文本序列的神经网络。

**类比：想象一个翻译员在逐字阅读一本书**

```
输入: "I love Beijing Tiananmen"

RNN 逐字处理：
  "I" → "love" → "Beijing" → "Tiananmen"
  1ms    2ms      3ms         4ms

总耗时 = 1 + 2 + 3 + 4 = 10ms
```

就像一个人排着队挨个读字，**必须读完前一个字，才能去读下一个**。

**问题 1：速度慢** — 每个字必须等前一个字处理完

**问题 2：健忘** — 读到句子末尾时，开头的内容已经模糊了

> 💡 **记忆要点**
> - RNN 像一个人逐字阅读，**串行处理，速度慢**
> - 距离越远的信息越容易被遗忘，这叫做"**长距离依赖问题**"

### v2.0 — LSTM：加了"笔记本"的翻译员

LSTM（Long Short-Term Memory）是 RNN 的升级版，加了"笔记本"来解决健忘问题。

LSTM 通过三个"门"来控制信息的保留和遗忘：

- **遗忘门**：决定忘掉哪些旧信息
- **输入门**：决定记住哪些新信息
- **输出门**：决定输出哪些信息

就像三个开关：清理旧数据、保存新数据、决定展示什么。

**解决的问题：** 记忆力变好了。**没解决的问题：** 仍然是串行处理，速度依然慢！

> 💡 **记忆要点**
> - LSTM 用"门机制"解决了记忆问题，但本质上仍然是**串行处理**
> - 类比：给翻译员配了笔记本，记忆力好了，但他还是一个人在干活

### v3.0 — Attention：让所有人同时工作

2017 年，Google 发表了划时代的论文 "Attention Is All You Need"。

**类比：从"一个翻译员"变成"一个翻译团队"**

```
RNN/LSTM（一个翻译员）:
  逐字阅读: "I" → "love" → "Beijing" → "Tiananmen"
  耗时: 4 个单位

Attention（一个翻译团队）:
  所有成员同时阅读整个句子！

  成员 A 读 "I"         |
  成员 B 读 "love"      | 全部同时进行！
  成员 C 读 "Beijing"   |
  成员 D 读 "Tiananmen" |
  耗时: 1 个单位
```

**核心思想：每个字都能直接"看到"所有其他字**，不需要排队传递信息。

> 💡 **记忆要点**
> - Attention 解决了两大问题：**并行计算**（速度快）+ **长距离依赖**（不再健忘）
> - 从一个人逐字阅读，变成一个团队同时阅读整篇文章

### v4.0 — Transformer：集大成者

Transformer 把 Attention 机制作为核心，搭建出完整架构。

**类比：Transformer = 一栋写字楼**，每层包含 Self-Attention（团队讨论）和 FFN（独立思考）。层数越多，理解越深入。

```
第  1 层：基础语法（主-谓-宾）
第  4 层：短语搭配（"吃 + 面条" 合理）
第  8 层：语义关系
第 12 层：深层含义理解
```

| 特性 | RNN | LSTM | Attention | Transformer |
|------|-----|------|-----------|-------------|
| 处理方式 | 串行 | 串行 | 并行 | 并行 |
| 记忆力 | 差 | 好 | 好 | 好 |
| 速度 | 慢 | 慢 | 快 | 快 |
| 长文本 | 差 | 一般 | 好 | 非常好 |

> 💡 **记忆要点**
> - Transformer = Attention + FFN，堆叠多层形成深度网络
> - 2017 年之后，几乎所有大模型（GPT、BERT、T5、LLaMA）都基于 Transformer

## 2. GPT vs BERT vs T5 的区别

有了 Transformer 这个基础架构后，不同团队用它构建了不同风格的模型。

![GPT vs BERT vs T5 对比](images/gpt_bert_t5.png)

*图2：三种 Transformer 变体架构*

### 类比：三种不同的考试方式

**GPT — 续写作文**：只能看到前面已经写过的字，一个字一个字往后写。擅长写文章、对话、代码生成。

**BERT — 完形填空**：能看到前后所有的字，填入被遮住的词。擅长理解文本、分类、问答。

**T5 — 翻译/转换**：先完整理解输入，再从头生成输出。擅长翻译、摘要、格式转换。

### 详细对比

| 维度 | GPT | BERT | T5 |
|------|-----|------|-----|
| 架构 | 仅解码器 | 仅编码器 | 编码器-解码器 |
| 注意力方向 | 从左到右 | 双向 | 两者兼有 |
| 训练方式 | 预测下一个词 | 填空 | 文本到文本 |
| 核心能力 | **生成** | **理解** | **转换** |
| 代表模型 | GPT-2/3/4 | BERT, RoBERTa | T5, FLAN-T5 |
| 典型用途 | 对话、写作、编程 | 搜索、分类 | 翻译、摘要 |

### 谁赢了？

GPT 成为主流的原因：**用户需要的是"生成"（帮我写文章、写代码、回答问题），而不只是"理解"（判断情感是正面还是负面）**。当前几乎所有对话式 AI 都采用 GPT 风格的 Decoder-Only 架构。

> 💡 **记忆要点**
> - **GPT = 作家**（生成），**BERT = 编辑**（理解），**T5 = 翻译官**（转换）
> - GPT 成为主流的原因：**用户需要"生成"**

### 理解 Tokenizer

在把文本交给模型之前，Tokenizer 会先把文本转换成数字序列。GPT-2 使用 BPE (Byte Pair Encoding) 分词方法。以 `Ġ` 开头的 token 表示前面有一个空格。

````python
# 演示 Tokenizer 的编码和解码
text = "Hello, I am a language model."

# 编码：文本 -> token ID
token_ids = tokenizer.encode(text)
# 获取每个 token 的文本表示
tokens = tokenizer.convert_ids_to_tokens(token_ids)

print(f"原始文本: {text}")
print(f"Token IDs: {token_ids}")
print(f"Tokens:    {tokens}")
print(f"Token 数量: {len(token_ids)}")
print()

# 解码：token ID -> 文本
decoded = tokenizer.decode(token_ids)
print(f"解码结果: {decoded}")
print()

# 词汇表大小
print(f"GPT-2 词汇表大小: {tokenizer.vocab_size} 个 token")
````

## 3. 探索模型内部结构

GPT-2 Small 的结构：

```
输入 Token IDs
    |
    v

|  +-- Feed Forward      |
|  +-- Layer Norm        |
+------------------------+
    |

    |
    v
输出：下一个 token 的概率分布
```

````python
# --- 5.1 打印模型的高层结构 ---
print("=" * 60)
print("GPT-2 Small 模型结构")
print("=" * 60)
print(model)
````

````python
# --- 5.2 统计每一层的参数量 ---
print("=" * 60)
print("各模块参数量统计")
print("=" * 60)

# 按模块分组统计参数
module_params = {}
for name, param in model.named_parameters():
    # 取顶层模块名（如 transformer.wte, transformer.h.0 等）
    parts = name.split(".")
    if len(parts) >= 3 and parts[1] == "h":
        # Transformer 层：归类为 "transformer.h.N"
        module_name = f"{parts[0]}.{parts[1]}.{parts[2]}"
    elif len(parts) >= 2:
        module_name = f"{parts[0]}.{parts[1]}"
    else:
        module_name = parts[0]
    
    if module_name not in module_params:
        module_params[module_name] = 0
    module_params[module_name] += param.numel()

# 按参数量排序并显示
print(f"\n{'模块名称':<30} {'参数量':>12} {'占比':>8}")
print("-" * 52)

total = sum(module_params.values())
for name, count in sorted(module_params.items()):
    pct = count / total * 100
    print(f"{name:<30} {count:>12,} {pct:>7.1f}%")

print("-" * 52)
print(f"{'总计':<30} {total:>12,} {'100.0%':>8}")
print(f"\n约 {total / 1e6:.1f}M 参数")
````

## 4. 大模型的核心能力

大模型不只是"更大的语言模型"，它展现出了一些小模型完全不具备的能力。

### 能力 1：In-Context Learning（上下文学习）

不需要重新训练，给几个例子就能学会新任务！

```
传统机器学习：
  收集 10,000 条标注数据 → 训练模型 → 部署上线 → 耗时数周

大模型方法：
  在提示词里给几个示例：
    "good" → 正面
    "terrible" → 负面
    "so-so" → 中性

    分类: "this movie is amazing" → ?
    大模型回答: "正面"

  耗时：几秒钟，无需额外训练！
```

### 能力 2：Chain-of-Thought（思维链推理）

大模型能像人一样"一步步思考"来解决复杂问题，而不是直接猜答案。

### 能力 3：Instruction Following（指令遵循）

能理解并执行各种复杂的人类指令，包括写代码、翻译、总结等。

### 能力 4：多任务通用能力

一个模型搞定所有 NLP 任务（翻译、摘要、分类、问答、编程...），不需要每个任务单独训练一个模型。

### 能力 5：Emergent Abilities（涌现能力）

模型大到一定程度后，突然"开窍"，出现了训练时没有明确教过的能力（数学推理、代码生成、多语言翻译等）。就像水烧到 100°C 突然沸腾一样。

> 💡 **记忆要点**
> - 5 大核心能力：**上下文学习、思维链推理、指令遵循、多任务通用、涌现能力**
> - 最神奇的是"涌现能力"——模型没有被明确教过，但大到一定程度后自己学会了

## 5. 文本生成：体验 GPT 的核心能力

语言模型的核心任务是**预测下一个 token**。给定一段提示词，模型不断预测下一个最可能的 token 来"续写"文本。不同的解码策略会显著影响生成结果：

| 策略 | 描述 | 特点 |
|------|------|------|
| **Greedy（贪心）** | 每次选概率最高的 token | 结果确定，但容易重复 |
| **Sampling（采样）** | 从概率分布中随机采样 | 多样性好，但可能不连贯 |
| **Top-k** | 只从概率最高的 k 个 token 中采样 | 平衡多样性和质量 |
| **Top-p** | 从累积概率 ≥ p 的最小集合中采样 | 自适应候选集大小 |

**Temperature** 控制随机程度：低温（0.3）更保守，高温（1.5）更冒险。

````python
def generate_text(prompt, max_new_tokens=50, **kwargs):
    """使用 GPT-2 生成文本的辅助函数"""
    # 将提示词编码为 token ID，并移到正确的设备
    inputs = tokenizer.encode(prompt, return_tensors="pt").to(device)
    
    # 生成文本（关闭梯度计算以节省显存）
    with torch.no_grad():
        outputs = model.generate(
            inputs,
            max_new_tokens=max_new_tokens,
            pad_token_id=tokenizer.eos_token_id,  # 避免 padding 警告
            **kwargs
        )
    
    # 解码生成的 token ID 为文本
    generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated


# --- 试试几个不同的提示词 ---
prompts = [
    "Once upon a time",
    "The meaning of life is",
    "Artificial intelligence will",
]

print("=" * 60)
print("贪心解码 (Greedy Decoding) 文本生成")
print("=" * 60)

for prompt in prompts:
    result = generate_text(prompt, max_new_tokens=50, do_sample=False)
    print(f"\n提示词: \"{prompt}\"")
    print(f"生成结果:\n{result}")
    print("-" * 60)
````

贪心解码每次选概率最高的 token，容易陷入循环。让我们对比不同策略：

````python
prompt = "The future of artificial intelligence is"

print("=" * 60)
print(f"提示词: \"{prompt}\"")
print("=" * 60)

# --- 策略1: 贪心解码（每次选概率最大的 token）---
print("\n【策略1】贪心解码 (Greedy):")
result_greedy = generate_text(prompt, max_new_tokens=60, do_sample=False)
print(result_greedy)

# --- 策略2: 采样 + 低温度（较保守）---
print("\n【策略2】采样 (Temperature=0.3, 低温 = 保守):")
result_low_temp = generate_text(
    prompt, max_new_tokens=60,
    do_sample=True, temperature=0.3
)
print(result_low_temp)

# --- 策略3: 采样 + 高温度（更有创意）---
print("\n【策略3】采样 (Temperature=1.2, 高温 = 冒险):")
result_high_temp = generate_text(
    prompt, max_new_tokens=60,
    do_sample=True, temperature=1.2
)
print(result_high_temp)

# --- 策略4: Top-k 采样 ---
print("\n【策略4】Top-k 采样 (k=50):")
result_topk = generate_text(
    prompt, max_new_tokens=60,
    do_sample=True, top_k=50, temperature=0.7
)
print(result_topk)

# --- 策略5: Top-p (Nucleus) 采样 ---
print("\n【策略5】Top-p 采样 (p=0.9):")
result_topp = generate_text(
    prompt, max_new_tokens=60,
    do_sample=True, top_p=0.9, temperature=0.7
)
print(result_topp)
````

**观察要点**：
1. 贪心解码结果确定但容易重复
2. 低温接近贪心，高温更有"创意"但可能不连贯
3. Top-k/Top-p 在多样性和质量间取得平衡

实际应用中，ChatGPT 等模型通常使用 Top-p + Temperature 的组合。

````python
# --- 多次运行采样，观察结果的多样性 ---
print("=" * 60)
print("同一个提示词，采样3次的不同结果")
print("=" * 60)

prompt = "In a world where robots"
for i in range(3):
    result = generate_text(
        prompt, max_new_tokens=40,
        do_sample=True, top_p=0.9, temperature=0.8
    )
    print(f"\n第 {i+1} 次: {result}")
````

## 6. 为什么大模型如此强大

### 核心原因：Scaling Law（规模定律）

2020 年，OpenAI 发现了一个惊人的规律：**模型的能力与三个因素的关系极其简单且可预测。**

![Scaling Law](images/scaling_law.png)

*图3：Scaling Law（规模定律）*

**Performance ∝ f(N, D, C)**
- **N** = 模型参数量（模型有多大）
- **D** = 训练数据量（看了多少书）
- **C** = 计算量（用了多少算力）

三者越大，模型性能越好，而且是可预测的幂律关系。

**类比：培养一个学生**
- N（大脑容量）：1B ≈ 小学生，10B ≈ 高中生，100B+ ≈ 博士生
- D（读了多少书）：1GB ≈ 几本书，1TB ≈ 一个图书馆，10TB+ ≈ 互联网
- C（学习时间）：越长理解越深入

关键发现：三者需要**同步增长**。只增大模型不增加数据会过拟合（天才只读3本书），只增加数据不增大模型会装不下（让小学生读博士论文）。

### 三个阶段：从"堆参数"到"会思考"

![训练三阶段](images/training_pipeline.png)

*图4：大模型训练的三个阶段*

1. **预训练（Pre-training）** — 读遍互联网文本，学会语言和世界知识。代价：数百万美元
2. **监督微调（SFT）** — 用高质量问答对训练，从"知识库"变成"助手"。代价：数万美元
3. **人类反馈强化学习（RLHF）** — 让回答更符合人类偏好，更有帮助、更安全。代价：数十万美元

### 为什么"预测下一个词"就能产生智能？

这是大模型最反直觉的地方：

```
训练目标看似极其简单：
  给定 "今天天气" → 预测 "不错"

但为了做好这件事，模型必须学会：
  - 语法规则
  - 常识（天气可以是晴天、雨天、阴天...）
  - 逻辑（如果前文提到下雨，天气大概不好）
  - 世界知识（北京冬天通常很冷）
  - 数学（预测正确的计算结果）
  - 编程（预测正确的代码语法）

"预测下一个词" —— 这个简单的目标，
配合足够大的模型和足够多的数据，
迫使模型理解了整个世界。
```

类比：一个学生只是为了通过"完型填空"考试，但为了做好这项考试，他不得不把语文、数学、历史、地理全部学会——因为完型填空的内容可能涉及任何领域！

> 💡 **记忆要点**

## 7. 观察"预测下一个 token"的实际过程

我们刚才讲了"预测下一个词"是大模型的核心。让我们手动看看模型对下一个 token 的预测概率分布：

````python
# 给定一段文本，查看模型预测的 Top-10 下一个 token
text = "The capital of France is"

# 编码输入
input_ids = tokenizer.encode(text, return_tensors="pt").to(device)

# 前向传播，获取模型输出
with torch.no_grad():
    outputs = model(input_ids)

# outputs.logits 的形状: [batch_size, seq_len, vocab_size]
# 我们只关心最后一个位置的预测（即下一个 token）
next_token_logits = outputs.logits[0, -1, :]  # shape: [50257]

# 转换为概率分布
probs = torch.softmax(next_token_logits, dim=-1)

# 取概率最高的 10 个 token
top_probs, top_indices = torch.topk(probs, 10)

print(f'输入: "{text}"')
print(f"\n模型预测的 Top-10 下一个 token:")
print(f"{'排名':<6} {'Token':<15} {'概率':>10}")
print("-" * 35)
for i, (prob, idx) in enumerate(zip(top_probs, top_indices)):
    token_text = tokenizer.decode(idx.item())
    print(f"{i+1:<6} {repr(token_text):<15} {prob.item():>10.4f}")
````

**解读**: 模型大概率会预测 " Paris"。这说明 GPT-2 在预训练过程中已经学到了世界知识——它知道法国的首都是巴黎。

这就是"预测下一个词"的魔力：为了准确预测，模型必须理解整个世界。
